## Immunity info

I want to get information about the date, when vaccination programs can be closed.  
To do that, it is necessary to determine the threshold, after which, the vaccinations can be stopped.  
The threshold, in a given population, is the point where the disease reaches a steady state, which means that the infection level is neither growing nor declining exponentially.  
The threshold is evaluated using the formula below.

$ p_c = 1 - \frac 1 R_0\ $

 
where $ R_0 - $  the basic reproduction number, the average number of new infections caused by each case in a population where each individual is equally likely to come into contact with any other susceptible individual in the population;  
$ P_c - $  the critical proportion of the population needed to be immune to stop the transmission of disease, which is the same as the "herd immunity threshold" (HIT).  
Information about herd immunity thresholds was found <a href="https://en.wikipedia.org/wiki/Herd_immunity#Theoretical_basis">here</a>.

For this project, the lowest threshold of 75% for the alpha variant of COVID-19 was calculated.  

However, it is important to point out that a lot of people could get immunity from COVID-19 after having had the disease.  
That could have lead to lower percentage of the population needed for reaching herd immunity threshold.  
Unfortunately, in this project this factor was not taken into account.
 

## Imports

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as sps
from datetime import timedelta, date, datetime
from tqdm import tqdm

from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
from pmdarima import auto_arima
import warnings

warnings.filterwarnings("ignore")
```

## Data

```python
# The dataframe is available on
# https://github.com/owid/covid-19-data
# https://www.kaggle.com/gpreda/covid-world-vaccination-progress
# The original dataframe contains 104214 observations of 16 parameters.
# For this project only 8 parameters were subsetted and used
```

```python
# Download data from Github
data = pd.read_parquet("../data/processed_data.parquet")
```

## Countries achieved heard immunity before 2022

```python
# countries achieved heard immunity before 2022
heard_immunity_before_2022 = (
    data.groupby("country")
    .filter(lambda x: x["people_fully_vaccinated_per_hundred"].max() >= 75)
    .groupby("country")
    .people_fully_vaccinated_per_hundred.agg(["max"])
)
heard_immunity_before_2022.shape
```

```python
heard_immunity_before_2022
```

## Countries did not achieve heard immunity before 2022

```python
train_data = data.loc[
    ~data["country"].isin(heard_immunity_before_2022.index.to_list()),
]
```

```python
countries = train_data["country"].unique().tolist()
len(countries)
```

## Sigmoid-curve fitting

The sigmoid-curve was used



$ f(x) = \frac L {1+{e}^{-k * (x-x_0)}} \ $

```python
# define sigmoidal curve
def func(x, L, k, x0):
    return L / (1 + np.exp(-k * (x - x0)))


# define reverse sigmoidal curve
def func_predict(y, L, k, x0):
    return -np.log(L / y - 1) / k + x0
```

```python
predictions = []
r2_scores = []
for country in tqdm(countries):
    subset = train_data.loc[
        (train_data["country"] == country),
        ["date", "people_fully_vaccinated_per_hundred"],
    ]
    x = list(range(subset.shape[0]))
    y = subset.people_fully_vaccinated_per_hundred
    try:
        popt, pcov = curve_fit(func, x, y)
    except RuntimeError:
        r2_scores.append(0)
        predictions.append(0)
        continue
    r2_scores.append(r2_score(y, func(x, *popt).round(3)))
    try:
        date = datetime.strptime((np.array(subset.date)[0]), "%Y-%m-%d") + timedelta(
            days=func_predict(75, *popt).round(0)
        )
        predictions.append(date.strftime("%d %b %Y"))
    except ValueError:
        predictions.append(0)
```

```python
r2_scores = np.array(r2_scores).round(3).tolist()
```

```python
results = pd.DataFrame(
    {
        "predictions": predictions,
        "r2_scores": r2_scores,
        "countries": countries,
    }
)
```

```python
results[results["predictions"] != 0]
```

```python
# countries that achieved herd immunity

continents = results[results["predictions"] != 0].countries.tolist()
fig, ax = plt.subplots(2, 3, figsize=(18, 8))

for i, continent in enumerate(continents):
    j = i % 3
    i = i // 3

    sample = data[data["country"] == continent]

    x = list(range(sample.shape[0]))
    y = sample.people_fully_vaccinated_per_hundred
    popt, pcov = curve_fit(func, x, y)

    ax[i, j].plot(
        pd.to_datetime(sample["date"]),
        sample["people_fully_vaccinated_per_hundred"],
        label="raw",
    )
    ax[i, j].plot(
        pd.to_datetime(sample["date"]),
        func(x, *popt).round(3),
        label="curve",
    )
    ax[i, j].grid()
    ax[i, j].legend()
    ax[i, j].set_title(continent)
    ax[i, j].set_xlabel("date")
    ax[i, j].tick_params(axis="x", labelrotation=45)
    ax[i, j].set_ylabel("fully vaccinated per hundred")

fig.tight_layout()
plt.show()
```

```python
# countries that not achieved herd immunity

continents = results[results["predictions"] == 0].tail(6).countries.tolist()
fig, ax = plt.subplots(2, 3, figsize=(18, 8))

for i, continent in enumerate(continents):
    j = i % 3
    i = i // 3

    sample = data[data["country"] == continent]

    x = list(range(sample.shape[0]))
    y = sample.people_fully_vaccinated_per_hundred
    popt, pcov = curve_fit(func, x, y)

    ax[i, j].plot(
        pd.to_datetime(sample["date"]),
        sample["people_fully_vaccinated_per_hundred"],
        label="raw",
    )
    ax[i, j].plot(
        pd.to_datetime(sample["date"]),
        func(x, *popt).round(3),
        label="curve",
    )
    ax[i, j].grid()
    ax[i, j].legend()
    ax[i, j].set_title(continent)
    ax[i, j].set_xlabel("date")
    ax[i, j].tick_params(axis="x", labelrotation=45)
    ax[i, j].set_ylabel("fully vaccinated per hundred")

fig.tight_layout()
plt.show()
```

## Trend prediction with ARIMA

```python
trends = pd.DataFrame()

for country in tqdm(countries):
    subset = train_data.loc[
        (train_data["country"] == country),
        ["date", "people_fully_vaccinated_per_hundred"],
    ]
    x = list(range(subset.shape[0]))
    y = subset.people_fully_vaccinated_per_hundred
    arima_model = auto_arima(
        y,
        seasonal=False,
        error_action="ignore",
        n_jobs=8,
        stationary=False,
        # trace=True,
        random_state=10,
    )

    predictions = arima_model.predict(n_periods=365)
    dates = pd.date_range(
        pd.to_datetime(sample.date.max()) + timedelta(days=1),
        pd.to_datetime(sample.date.max()) + timedelta(days=365),
    )
    subset_reult = pd.DataFrame(
        {
            "date": dates,
            "predictions": predictions,
            "country": country,
        }
    )
    trends = pd.concat([trends, subset_reult])
```

```python
# predicted trends for six countries

countries = trends["country"].unique().tolist()[:6]
# results[results["predictions"] == 0].tail(6).countries.tolist()
fig, ax = plt.subplots(2, 3, figsize=(18, 8))

for i, continent in enumerate(countries):
    j = i % 3
    i = i // 3

    sample = data[data["country"] == continent]
    trend_sample = trends[trends["country"] == continent]

    x = list(range(sample.shape[0]))
    y = sample.people_fully_vaccinated_per_hundred
    popt, pcov = curve_fit(func, x, y)

    ax[i, j].plot(
        pd.to_datetime(sample["date"]),
        sample["people_fully_vaccinated_per_hundred"],
        label="raw",
    )
    ax[i, j].plot(
        trend_sample.date,
        trend_sample.predictions,
        label="curve",
    )
    ax[i, j].grid()
    ax[i, j].legend()
    ax[i, j].set_title(continent)
    ax[i, j].set_xlabel("date")
    ax[i, j].tick_params(axis="x", labelrotation=45)
    ax[i, j].set_ylabel("fully vaccinated per hundred")

fig.tight_layout()
plt.show()
```

## Results

- The countries that reached the threshold of herd immunity were found.   
    - 19 countries reached herd immunity before 2022
    - 40 countries are expected to reach herd immunity before 2023
- For those countries that can reach this threshold, the date of the threshold attainment was calculated.    

- The vaccination rates were examined, cleaned and plotted for each country and continent.
- The vaccination trends were obtained. 
    - Two approaches were implemented: sigmoid-like curve fitting and ARIMA.
    - Sigmoid-like curve fitting perfectly explained the historical trend, however was "" for trend prediction
    - ARIMA was perfect for recent data trend extrapolation, however it was impossible to model saturation and trend was linear.
    - Other counties were not analyzed due to lack of data.

- Out of ~200 countries only 59 showed accomplishment of reaching the status of 75% immunized population, which is the lowest threshold for herd immunity for alpha variant of COVID-19.
- Other countries cannot surpass the threshold of herd immunity.
